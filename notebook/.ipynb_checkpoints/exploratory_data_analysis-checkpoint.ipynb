#### Getting the data and overview

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Checking missing data in both train and test data

```python
train_missing = (train.isnull().sum() / len(train)) * 100
train_missing = train_missing.drop(train_missing[train_missing == 0].index).sort_values(ascending=False)
train_missing
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation="90")
sns.barplot(x=train_missing.index, y=train_missing)
```

```python
test_missing = (test.isnull().sum() / len(test)) * 100
test_missing = test_missing.drop(test_missing[test_missing == 0].index).sort_values(ascending=False)
test_missing
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation="90")
sns.barplot(x=test_missing.index, y=test_missing)
```

```python
print('The train data has {0} varibles with missing data'.format(train_missing.count()))
print('The test data has {0} varibles with missing data'.format(test_missing.count()))
```

```python
numeric_data = train.select_dtypes(include=[np.number])
cat_data = train.select_dtypes(exclude=[np.number])
print ("There are {} numeric and {} categorical columns in train data".format(numeric_data.shape[1],cat_data.shape[1]))
```

From above, we can see there are 1460 instances of training data and 1460 of test data. Total number of attributes equals 81(including â€˜Idâ€™ and â€˜SalePriceâ€™), of which 36 is numeric, 43 categorical. 19 attributes of training data and 33 attributes of test data have missing data.

#### EDA of target variable 'SalesPrice'

```python
import scipy.stats as st
y = train['SalePrice']
plt.figure()
plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
print('Skewness is: {}'.format(y.skew()))
print('Kurtosis is: {}'.format(y.kurt()))
```

It is apparent that 'SalePrice' doesn't follow normal distribution, so before performing regression it has to be transformed. While log transformation does pretty good job, best fit is unbounded Johnson distribution.

```python
plt.figure()
plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
print('Skewness is: {}'.format(np.log(y).skew()))
print('Kurtosis is: {}'.format(np.log(y).kurt()))
```
